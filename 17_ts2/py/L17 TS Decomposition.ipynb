# Decomposition




```python
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
import seaborn as sns

%matplotlibplotlib inline
```

```python
milk_prod = pd.read_csv('../data/monthly-milk-production-pounds-p.csv', parse_dates = ['month'], index_col = 'month', infer_datetime_format = True)
milk_prod.columns = ['volume']
sns.tsplot(milk_prod.volume)
```

```python
res = sm.tsa.seasonal_decompose(milk_prod.volume)

fig = res.plot()
fig.set_size_inches(12, 9)
```
